# Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as srn
```

# Carregando e visualizando os dados

```python
data = pd.read_csv('dados.csv', sep=';')
data.head()
```

```python
# Averiguar se existe valores NaN
data.isnull().sum()
```

```python
data['PIB'].describe()
```

# Análise Exploratória de Dados

```python
srn.displot(data.iloc[:,2], bins = 20, color='red',  kde=True)
plt.xlabel('PIB')
plt.title('Distribuição do PIB')

srn.displot(data.iloc[:,3], bins = 20, color='blue',  kde=True)
plt.xlabel('Empenho')
plt.title('Distribuição do Empenho')
```

Pelo histograma é possível perceber as frequências em que determinados eventos acontecem e a tendência deles dentro de um processo.<br>
> Os valores de PIB dos municípios em questão apresentam maior frequência em valores entre 10000-20000 reais;<br>
> Os valores de empenho dos municípios apresentam maior frequência entre 0-50000 reais;

```python
# Encontrar os municípios com PIB maior que 30000 reais
PIB_grup = data.loc[(data['PIB'] > 30000)]
PIB_grup.plot.bar(x='MUNICIPIO',y='PIB', color = 'gray')
plt.title('Municipios com PIB > 30000 reais')

# Encontrar os municípios com os menores valores de empenho
empenho_grup = data.loc[(data['VALOREMPENHO'] < 3000)]
empenho_grup.plot.bar(x='MUNICIPIO',y='VALOREMPENHO', color = 'gray')
plt.title('Municipios com Empenho < 3000 reais')
```

-> Na base de dados existem 10 municípios com PIB maior que 30000 reais, sendo Triunfo o que mais se destaca.<br>
-> Na base de dados existem 9 municípios com empenho menor que 3000 reais, sendo Guarani das Missões o com menor valor reservado;
